```python
import os
import time
import random

# Initialize Keyboard pynput
from pynput.keyboard import Key ,Controller
keyboard = Controller()
# Initialize Mouse pynput
from pynput.mouse import Button , Controller
mouse = Controller()
# Import ImageGrab
from PIL import ImageGrab

# Create function to click
def tab():
    time.sleep(0.05)
    keyboard.press(Key.tab)
    keyboard.release(Key.tab)
    time.sleep(0.05)
def click(x, y, delay):
    mouse.position = (x, y)
    time.sleep(0.1)
    mouse.press(Button.left)
    mouse.release(Button.left)
    time.sleep(delay)
def right_click(x, y, delay):
    mouse.position = (x, y)
    time.sleep(0.1)
    mouse.press(Button.right)
    mouse.release(Button.right)
    time.sleep(delay)
# Create function to turn on/off do not disturb
def do_not_disturb(status):
    if status == "on":
        time.sleep(2.0)
        click(27.11328125, 11.875, 0.1) # Apple logo
        mouse.move(0, 50)
        time.sleep(0.1)
        mouse.click(Button.left, 1) # Sys pref
        mouse.move(525, -20)
        time.sleep(1.00)
        mouse.click(Button.left, 1)
        keyboard.press(Key.backspace) # delete search
        keyboard.release(Key.backspace)
        time.sleep(0.1)
        keyboard.press(Key.backspace)
        keyboard.release(Key.backspace)
        keyboard.type("Do Not Disturb")
        time.sleep(0.1)
        keyboard.press(Key.enter)
        keyboard.release(Key.enter)
        time.sleep(0.5)
        keyboard.press(Key.enter)
        keyboard.release(Key.enter)
        time.sleep(3.0)
        keyboard.press(Key.backspace)
        keyboard.release(Key.backspace)
        tab()
        tab()
        keyboard.type("4")
        tab()
        keyboard.type("00")
        tab()
        keyboard.type("A")
        tab()
        keyboard.type("3")
        tab()
        keyboard.type("59")
        tab()
        keyboard.type("A")
        time.sleep(0.1)
        mouse.move(-250, 230)
        mouse.click(Button.left, 1)
        time.sleep(0.5)
        click(19.26953125, 42.296875, 0)
    elif status == "off":
        click(1411.1640625, 13.01953125, 1) # Open notification center
        mouse.move(0,100) # Move cursor into section
        time.sleep(0.1)
        mouse.scroll(0, 50) # Scroll up to do not disturb
        time.sleep(0.1)
        click(1391.5625, 149.03125, 0.25) # Switch do not disturb setting
        click(1411.1640625, 13.01953125, 0.25) # Close notification center
    
# For Mac: System Preferences > Security & Privacy > Accessibility > Allow Terminal

# Turn on Do Not Disturb (notifications can mess with clicks while AFK)
do_not_disturb("on")

#Opening Runescape, Waiting 15 Seconds for Application to Open (loading time varies on computer/network)
os.system("open -a Runescape")
time.sleep(15)

# LOG IN
click(665.04296875, 438.52734375, 0) # Password box
keyboard.type(password)
time.sleep(1.0)
click(710.59765625, 490.1875, 1) # Log In box
click(710.59765625, 490.1875, 15) # Log In box again (loading time varies on computer/network)
click(717.96484375, 593.24609375, 1.0) # Play Now
click(717.96484375, 560.24609375, 17.5) # Play Now (in case cache verification bar doesn't appear)

# Play
keyboard.press(Key.up)
time.sleep(0.5) # Adjust camera up
keyboard.release(Key.up)
mouse.scroll(0,-50) # Zoom out in case previous sign-in changed zoom
inventory_open = ImageGrab.grab(bbox = (2150, 1715, 2215, 1800)) # Grab image of bag icon being opened (for ending the program)
inventory_status = inventory_open # Set current status
inventory_empty = ImageGrab.grab(bbox = (2775, 1600, 2865, 1700)) # Grab image of bottom right log, detect if inventory is full. Syntax: bbox = (left, upper, right, lower)
no_trips = 0 # Set counter to zero

while inventory_status == inventory_open:    
    inventory_current = inventory_empty # Make images the same so harvesting process can begin
    
    while inventory_current == inventory_empty:
        click(720.7421875, 516.33984375, 5.0) # Chop tree, repeat every 5 seconds for tree to regrow
        inventory_current = ImageGrab.grab(bbox = (2775, 1600, 2865, 1700)) # Update inventory status
        
    right_click(956.3515625, 70.76171875, 0.25) # Move curser to deposit box
    mouse.move(0,2)
    time.sleep(0.15)
    mouse.move(0, 35)
    time.sleep(0.15)
    mouse.press(Button.left)
    mouse.release(Button.left)
    time.sleep(9.25) # Wait for user to reach deposit box
    inventory_status = ImageGrab.grab(bbox = (2150, 1715, 2215, 1800)) # Check to see if inventory is still open
    
    click(1305.24609375, 193.96875, 0) # Move back in front of tree
    if inventory_status == inventory_open:
        no_trips += 1
        if no_trips <= 5:
            keyboard.type("Number of trips: " + str(no_trips))
        elif no_trips <= 10:
            color = random.choice(["red:", "cyan:", "green:", "white:", "purple:"])
            keyboard.type(color + "Number of trips: " + str(no_trips))
        else:
            color = random.choice(["glow1:", "glow2:", "glow3:", "flash1:", "flash2:", "flash3:"])
            effect = random.choice(["wave:", "wave2:", "shake:", "scroll:", "slide:"])
            keyboard.type(color + effect + "Number of trips: " + str(no_trips))
    
    time.sleep(0.5)
    keyboard.press(Key.enter) # Send number of trips in chat
    keyboard.release(Key.enter)
    time.sleep(9) # wait for user to return to tree

# LOG OUT

while inventory_status != inventory_open: 
    click(1092.015625, 878.09375, 0.1) # Open inventory back up for next run
    mouse.move(-100,0) # Move cursor out of the way for image grab
    inventory_status = ImageGrab.grab(bbox = (2150, 1715, 2215, 1800)) # Keep running in case click is inconsistent
click(1430.5703125, 57.171875, 3) # Options Menu box
click(834.63671875, 567.09375, 0.75) # Logout box
click(834.63671875, 567.09375, 0.1) # Logout box again
do_not_disturb("off") # Switch notifications back on
click(13.828125, 34.9296875, 0) # Exit
print(str(no_trips) + " full trips made.\n" + str(no_trips * 28) + " logs cut.")
```
